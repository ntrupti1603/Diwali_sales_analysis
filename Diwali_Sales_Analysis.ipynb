```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv("/content/drive/MyDrive/Pune/Datasets/Diwali Sales Data.csv",encoding = 'unicode_escape') #utf-8
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#drop cols
df.drop(columns=['Status','unnamed1'],inplace =True)
```

```python
df.head()
```

```python
df.shape
```

```python
# df.dropna >>> To drop the Missing Values
#df.drop >>>  To drop the  Column
```

```python
df.isnull().mean()*100
```

```python
df.dropna(inplace = True)
```

```python
df.head()
```

```python
df['Amount'] = df['Amount'].astype('int')  #Updating The existing Col
```

```python
df.head()
```

```python
#rename the col (Marital_Status >>> )
df.rename(columns ={'Marital_Status':'Single'}, inplace = True)
```

```python
df.head()
```

```python
#Replacing the values inside a column
df['Single'] = df['Single'].replace(1,'No')
df['Single'] = df['Single'].replace(0,'Yes')
```

```python
df.head()
```

```python
df.shape
```

```python
#Duplicate
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace =True)
```

```python
sales_age =  df.groupby(['Age Group','Gender'])['Orders'].sum().sort_values(ascending = False).reset_index()
sales_age
```

```python
#Plotting using Seaborn
sns.barplot(data = sales_age , x = 'Age Group' , y = 'Orders',palette ='viridis',hue='Gender' )
```

```python
#Plotting using Matplotlib
plt.bar(sales_age['Age Group'] , sales_age['Orders'])
plt.xlabel("Age Group")
plt.ylabel("Orders")
```

```python
#Plotting using plotly
px.bar(sales_age , x = 'Age Group', y =  'Orders')
```

```python
#Amount spent by Age group
sales_age_amt =  df.groupby(['Age Group','Gender'])['Amount'].sum().sort_values(ascending = False).reset_index()
sales_age_amt
```

```python
#Plotting using Seaborn
sns.barplot(data = sales_age_amt , x = 'Age Group' , y = 'Amount',palette ='viridis',hue='Gender' )
```
